```python
## for the SPLIT functions, change the 3 to the number of groups you need (48)

pez.columns = np.arange(len(pez.columns))
drop1 = list(range(0,pez.shape[1],4))
df1 = pez.drop(drop1, axis=1)
df1.columns = np.arange(len(df1.columns))
drop2 = list(range(2,df1.shape[1],3))
df2 = df1.drop(drop2, axis=1).drop(0)
df2.columns = np.arange(len(df2.columns))
SPLIT1 = np.array_split(df2, 12, axis=1)#[1]
peztable = [item.dropna() for item in SPLIT1]

########################################################### LOOP FUNCTIONS ##########################################################################
psum = [item.sum() for item in peztable]
pmean = [item.mean() for item in peztable]
pcount = [item.count() for item in peztable]
psize = [item.median() for item in peztable]

pfill = pd.DataFrame(psum).fillna(0)
pfill1 = pd.DataFrame(pmean).fillna(0)
pfill2 = pd.DataFrame(pcount).fillna(0)
pfill3 = pd.DataFrame(psize).fillna(0)

pezsums = pd.DataFrame(pfill.iloc[0:,0::2].sum(0)).set_axis(['pEz Sums'], axis='columns').reset_index().iloc[0:, 1:]
pezmeans = pd.DataFrame(pfill1.iloc[0:,1::2].sum(0)).set_axis(['pEz Means'], axis='columns').reset_index().iloc[0:, 1:]
pezcounts = pd.DataFrame(pfill2.iloc[0:,0::2].sum(0)).set_axis(['pEz Counts'], axis='columns').reset_index().iloc[0:, 1:]
pezsizes = pd.DataFrame(pfill3.iloc[0:,0::2].sum(0)).set_axis(['pEz Sizes'], axis='columns').reset_index().iloc[0:, 1:]
```

```python
#insert threshold table into concat function once all data collected

table1 = pd.concat((ttest, pezsums, pezmeans, pezcounts, pezsizes), axis=1)
table1
```
